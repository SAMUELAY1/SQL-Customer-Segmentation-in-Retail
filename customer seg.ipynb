```python
import pandas as pd 
```

```python
# reading cvs file
df = pd.read_csv('online_retail.csv')
df
```

```python
# cleaning dataset for project
```

```python
# droPing rows without customer id 
df = df.dropna(subset=['Customer ID'])
```

```python
# convert invoicedata to DATAfRAME 
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
df
```

```python
# Remove negative/zero quantities 
df = df[df['Quantity'] > 0]
df
```

```python
# create a lineTotal (quantity * unit price)
df['lineTotal'] = df['Quantity'] * df['Price']
df
```

```python
# trimming whitespace on text fields
df['Description'] = df['Description'].astype(str).str.strip()
df
```

```python
# change customerid to integer
df['Customer ID'] = df['Customer ID'].astype(int)
df
```

```python
# saving to csv
df.to_csv("online_retail_clean.csv")
```
